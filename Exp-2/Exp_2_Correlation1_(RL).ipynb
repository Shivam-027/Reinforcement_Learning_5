#**Correlation 1**


#Positive Correlation: if one data is increasing, then other data is also increasing


#Negative Correlation: if one data is decreasing, then other data is also decreasing.


# Zero Correlation: There is no relation between the data.

# **Positive correlation**

```python
import numpy as np
```

```python
np.random.seed(1)
```

```python
#1000 random integers between 0 and 50
x = np.random.randint(0, 50, 1000)
print(x)
```

```python
#positive correlation with some noise
y = x + np.random.normal(0, 10, 1000)
print(y)

```

```python
np.corrcoef(x,y)
```

This correlation is 0.815, a strong positive correlation

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')

plt.scatter(x,y)
plt.show
```

This is a positive correlation, as the graph is increasing and is in a straight line.

# **Negative Correlation**

```python
#1000 random integers between 0 and 50
a = np.random.randint(0, 50, 1000)
print(a)
```

```python
#positive correlation with some noise
b = 100 - a + np.random.normal(0, 5, 1000)
print(b)
```

```python
np.corrcoef(a,b)
```

This correlation is -0.94, a strong negative correlation

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')

plt.scatter(a,b)
plt.show
```

This is a negative correlation, as the graph is decreasing and is in a straight line.

# **Checking Correlation with three variables**

```python
#1000 random integers between 0 and 50
e = np.random.randint(0, 50, 1000)
print(e)

#positive correlation with some noise
f = e + np.random.normal(0, 10, 1000)
print(f)

g = np.random.randint(0, 10, 1000)
print(g)
```

```python
np.corrcoef(e,g)
```

```python
np.corrcoef(f,g)
```

```python
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline
matplotlib.style.use('ggplot')

plt.scatter(e,f,g)
plt.show
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Create a 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(e, f, g)

ax.set_xlabel('E')
ax.set_ylabel('F')
ax.set_zlabel('G')

plt.show()
```

# **Correlation between mulitple columns using Dataframe**

```python
import pandas as pd

df = pd.DataFrame({'a': np.random.randint(0,50,1000)})
df['b'] = df['a'] + np.random.normal(0,10,1000)
df['c'] = 100 - df['a'] + np.random.normal(0,5,1000)
df['d'] = np.random.normal(0,50,1000)

df.corr()
```

**This is the correlation between 4 columns**

```python
pd.plotting.scatter_matrix(df, figsize=(6,6))
plt.show()
```

```python
plt.matshow(df.corr())
plt.xticks(range(len(df.columns)), df.columns)
plt.yticks(range(len(df.columns)), df.columns)
plt.colorbar()
plt.show()
```

**Yellow indicates - 1 <br>
Light green indicates - positive correlation <br>
Dark purple indicates - Negative correlation <br>
blue indicates - No correlation <br>**
